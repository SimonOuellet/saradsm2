## Contrôler l'esthétisme et de la présentation d'un graphique Seaborn

Exemple pour créer un graphique à lignes simple avec des paramètres par défaut : 

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label='Groupe A')
plt.plot(x2, label='Groupe B')
plt.legend()
plt.show()
````

Pour activer les paramètres de configuration affichage par défaut de Seaborn, il faut simplement appeler la function **.set()**

````python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label='Group A')
plt.plot(x2, label='Group B')
plt.legend()
plt.show()
````

Seaborn organise ses paramètres en deux groupes. Le premier rassemble les paramètres pour l'esthétisme d'un graphique et le second fait varier la taille des éléments pour les adapter aux différents contextes.  

Pour contrôler le premier groupe qui permet de modifier le style et l'esthétisme d'un graphique, Seaborn fourni deux méthodes :  
**set_style(style, [rc])**  
et
**axes_style(style, [rc])**  

Paramètres :  
- **style**: un dictionnaire de paramètres ou le nom d'une des configurations pré-existente suivante: **darkgrid, whitegrid, dark, white, ou ticks**
- **rc**: un mapping de paramètres pour ré-écrire les valeurs proposé par défaut par Seaborn.

Pour ajuster plus finement les paramètres, il faut alors passer par les  **rc**

````python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("ticks")
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label='Groupe A')
plt.plot(x2, label='Groupe B')
plt.legend()
plt.show()
````

### Enlever les bordures d'un graphique
Parfois il est souhaitable d'enlever la bordure du haut et de droite. C'est possible en utilisant la fonction **.despine()**  
```
sns.despine(fig=None, ax=None, top=True, right=True,
                left=False, bottom=False
                offset=None, trim=False)
```

````python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("white")
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label='Group A')
plt.plot(x2, label='Group B')
sns.despine()
plt.legend()
plt.show()
````

## Contrôler la taille des éléments

Seaborn permet de contrôler la taille des éléments graphiques simplement à l'aide de la fonction **set_context()**  
Les contextes disponibles par défaut sont : **notebook, paper, talk,** et **poster**

````python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("poster")
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label='Group A')
plt.plot(x2, label='Group B')
plt.legend()
plt.show()
````

### Exemple plus complet avec des boites à moustaches

````python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
mydata = pd.read_csv("data/iq_scores.csv")
group_a = mydata[mydata.columns[0]].tolist()
group_b = mydata[mydata.columns[1]].tolist()
group_c = mydata[mydata.columns[2]].tolist()
group_d = mydata[mydata.columns[3]].tolist()
data = pd.DataFrame({'Groupes': ['Group A'] * len(group_a) + ['Group B'] * len(group_b) + ['Group C'] * len(group_c) + ['Group D'] * len(group_d),
                     'Resultats QI': group_a + group_b + group_c + group_d})
plt.figure(dpi=100)
# Utilise la préconfiguration de style 'whitegrid'
sns.set_style('whitegrid')
# Active le contexte 'notebook'
sns.set_context('talk')
# Créer la boite à moustache
sns.boxplot(data=data, x='Groupes', y='Resultats QI')
# Enlève les bordures
sns.despine(left=True, right=True, top=True)
# Ajoute le titre
plt.title('Résultats de QI des différents groupes')
# Affiche le graphique
plt.show()
````

## Palettes de couleurs
### Les palettes Catégoriques

Les couleurs sont une composantes importantes d'un graphique.  

Les palettes de couleurs catégoriques sont utiles pour les données catégoriques qui ne se suivent pas.  

Seaborn offre des palettes de couleurs pré-définies accessible via la fonction **color_palette()**.  
**seaborn.color_palette([palette], [n_colors], [desat])**   
paramètres:  
- **palette**(optionel): Nom de la palette ou None pour utiliser la palette actuelle.  
- **n_colors**(optionel): Nombre de couleur que la palette doit contenir. Si le nombre de couleur est plus élevé que la quantité de couleur disponible dans la palette, les couleurs vont cycler.  
- **desat**(optionnel): Proportion de désaturation de la palette (0.0 à 1.0)  

Il est possible d'appliquer la même palette pour tous vos graphiques en utilisant la fonction **set_palette()** qui utilise les mêmes paramètres que color_palette.  

Les 6 palettes par défaut sont : **deep, muted, bright, pastel, dark,** et **colorblind**

````python
sns.color_palette("bright", n_colors=20, desat=0.1)
````

````python
sns.color_palette("bright", n_colors=30, desat=0.5)
````

````python
sns.color_palette("bright", n_colors=10, desat=1)
````

````python
sns.color_palette("pastel")
````

````python
import seaborn as sns
palette1 = sns.color_palette("deep")
sns.palplot(palette1)
````

### Les palettes séquentielles

Les palettes de couleurs séquentielles sont comme leur nom le laisse supposer, utilent pour des données de type séquentiel.

**cubehelix_palette()** est un exemple de palette séquentielle pré-définie par Seaborn.  
La fonction permet d'indiquer la teinte (hue) de départ (**start** 0-3), le nombre de changement (helix) de teinte (**rot**) et le nombre de couleur souhaité.




````python
sns.cubehelix_palette(rot=0, n_colors=12)
````

````python
sns.cubehelix_palette(start=1, rot=0, n_colors=12)
````

````python
sns.cubehelix_palette(rot=1, n_colors=12)
````

````python
sns.cubehelix_palette(rot=2, n_colors=12)
````

Il est possible de créer sa propre palette à l'aide des fonctions **light_palette()** ou **dark_palette()**

````python
custom_palette2 = sns.light_palette("magenta")
sns.palplot(custom_palette2)
````

````python
custom_palette3 = sns.light_palette("magenta", reverse=True)
sns.palplot(custom_palette3)
````

Par défaut, il n'est retourné que la palette de couleur. Si vous avez besoin d'un objet de type **colormap** comme requis pour les graphiques de type *heatmap*, il faudra positionner le paramètre **as_cmap** à **True**.

````python
x = np.arange(25).reshape(5, 5)
ax = sns.heatmap(x, cmap=sns.cubehelix_palette(as_cmap=True))
````

### Palette de couleur divergeante
Ce sont des palettes pour les données ayant un milieu bien précis. L'emphase doit être bien concentrée dans les extrémités.  
C'est utile lorsque l'on veut démontrer un changement. Par exemple pour démontrer la croissance et décroissance d'une population.  

La palette **coolwarm** est un palette divergeante prédéfinis dans Seaborn.  
Il est possible de créer sa propre palette divergeante en utilisant la fonction **diverging_palette()**.  
Il est possible de passer deux teintes(hue) exprimées en degrés en paramètres.

````python
custom_palette4 = sns.color_palette("coolwarm", 12)
sns.palplot(custom_palette4)
````

````python
custom_palette5 = sns.diverging_palette(120, 300, n=12)
sns.palplot(custom_palette5)
````

### Exemple plus complet avec des graphiques de type heatmap
Un exemple qui démontre l'évolution des températures dans l'hémisphère nord.

````python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
data = pd.read_csv("data/northern_surface_temperature.csv", index_col=['Year'])
data = data.transpose()
heat_colormap = sns.diverging_palette(240, 15, s=99, as_cmap=True)
plt.figure(dpi=300)
sns.heatmap(data.iloc[:, ::5], cmap=heat_colormap, center=0)
plt.title("Changement des temperature depuis 1880 à 2015 (période de base 1951-1980)")
plt.savefig('temperature_change.png', dpi=300, bbox_inches='tight')
````

### Estimation par noyau
Pour connaitre la distribution de vos données, seaborn permet d'afficher un histogramme mono ou bi variable.   
Dans le cas d'un histogramme mono variable, il est possible d'afficher également l'estimation par densité du noyau (**[kde](https://fr.wikipedia.org/wiki/Estimation_par_noyau)**)

````python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('data/age_salary_hours.csv')
#plt.xlim(0,100)
sns.histplot(data.loc[:, 'Age'], kde=True, binrange=[0,100])

plt.xlabel('Age')
plt.ylabel('Density')
````

````python
sns.kdeplot(data.loc[:, 'Age'], fill=True)
plt.xlabel('Age')
plt.ylabel('Density')
````

### Graphique distribution bi-variables
Méthode 1 - utilisation de la fonction jointplot  
Cette méthode va générer un nuage de points et afficher sur les bordures un histograme permettant d'estimer la distribution de chaque axes.

````python
import pandas as pd
import seaborn as sns
data = pd.read_csv('data/age_salary_hours.csv')
sns.set(style="white")
sns.jointplot(x="Annual Salary", y="Age", data=data)
plt.show()
````

Il est également possible d'utiliser l'estimation par densité de noyau (KDE).

````python
import pandas as pd
import seaborn as sns
data = pd.read_csv('data/age_salary_hours.csv')
sns.jointplot(x='Annual Salary', y='Age', data=data, kind='kde', fill=True, xlim=(0, 500000), ylim=(0, 100))
plt.show()
````

### Visualiser les relations par paires
Afin de visualiser des relations par paires qui pourraient se trouver dans un jeu de données, seaborn possède la fonction **pairplot()** pour réaliser ce type de visualisation.  
Cette fonction crée une matrice dans laquelle les éléments hors diagonale démontrent la relation entre chaque paire de variables et les éléments diagonaux indiquent les distributions marginales.  

````python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('data/age_salary_hours.csv')
sns.set(style="ticks", color_codes=True)
g = sns.pairplot(data, hue='Education')
````

## Multi-graphiques avec Seaborn
### FacetGrid
FacetGrip permet de visualiser un certain type de graphique pour plusieurs variables séparément.  
Jusqu'à 3 dimensions peuvent être utilisées pour dessiner un FacetGrid: **row**, **col** et **hue**.  
Ces dimensions pourront être initialisées depuis le DataFrame mais les données devront être obligatoirement de type catégorielle ou discrète.  

Une fois initialisé, le FacetGrid n'affiche rien encore. Pour cela, il faut associer des données en utilisant la fonction **FacetGrid.map()**. Il est possible de passer n'importe quel type de fonction de graph et le nom des variables à utiliser depuis le dataframe.   
**FacetGrid.map(func, \*args, \*kwargs)**  
Paramètres:  
- **func**: le nom de la fonction graphique à utiliser.  
- **\*args**: le nom des variables à grapher. Les données de ces variables seront passé à la **func** dans l'ordre spécifié.  
- **\*kwargs**: liste d'arguements mot-clés qui seront passés à la **func**

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("data/salary.csv")[:1000]
g = sns.FacetGrid(data, col='District')
g.map(plt.scatter, 'Salary', 'Age')
````

### Graphique de régression
La régression est une technique pour estimer la relation entre une variable dépendante (la plupart du temps représentée sur l'axe des Y) et une variable indépendante (la plupart du temps représentée sur l'axe des x).  
Seaborn couvre un plus grand nombre de régression mais il ne sera traité ici que de la régression linéaire.  

La fonctioin **regplot()** permet de visualiser une relation de régression linéaire.


````python
import numpy as np
import seaborn as sns
x = np.arange(100)
# Une distribution normale avec une moyenne à 0 et écart-type de 5
y = x + np.random.normal(0, 5, size=100)
sns.set(style="ticks")
sns.regplot(x=x, y=y)
````
